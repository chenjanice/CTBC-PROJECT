```python
# -*- coding: utf-8 -*-
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re
import calendar

x=0
href_list =[]
title_list = []
year_list = []
month_list=[]
day_list=[]
for i in range(1,29):
    res = requests.get('https://www.cnbc.com/bill-ackman/?page='+str(i))
    if res.status_code == 200:
        content = res.content
        soup = BeautifulSoup(content, "html.parser")
        items = soup.findAll("div", {"class": "stories-lineup bigHeader"})
        for item in items:
            headline = item.findAll("div", {"class": "headline"})
            #print(headline)
            for tag in headline:
                t=tag.text
                t=t[30:]
                c=tag.select('a')
                h=c[0]['href']
                url='https://www.cnbc.com'+h
                if h[1:6] == 'video':
                    d=h[7:17]
                elif h[1:3] == 'id':
                    break
                else:
                    d=h[1:11]
                    
                y = d[0:4]
                m = d[5:7]
                a = d[8:10]
                        
            
                href_list.append(url)
                title_list.append(t)
                year_list.append(y)
                month_list.append(m)
                day_list.append(a)
                
                print(d)
                print(t[:60])
                print(url)
                print()
                x+=1

print(x)
```

```python
print(len(title_list))
print(len(href_list))
print(len(year_list))
print(len(month_list))
print(len(day_list))
```

```python
text_list = []
for i in range(len(href_list)):
    print(i,href_list[i])
    res2 = requests.get(href_list[i])
    if res2.status_code == 200:
        content = res2.content
        soup = BeautifulSoup(content, "html.parser")
        
        if href_list[i][21:26]=='video':
            items = soup.findAll("div", {"class": "ClipPlayer-clipPlayerIntroSummary"})
            try:
                s = items[0].text
            except:
                s = 'none'
            
        else:
            try:
                items = soup.findAll("div", {"class": "ArticleBody-articleBody"})
                for item in items:
                    s = " "
                    phases = item.findAll("p")
                    for phase in phases:
                        if phase is not None:
                            if phase.get("class") is None:
                                s += str(phase.text)                   
            except:
                s = 'none'
        
        text_list.append(s)
        print(len(text_list))
        print()

print('over')
```

```python
N_table = pd.DataFrame({'Title': title_list,
                        'Year': year_list,
                        'Month': month_list,
                        'Day':day_list,
                        'Content': text_list,
                        'Href_link': href_list})
N_table['Title'] = title_list
N_table['Year'] = year_list
N_table['Month'] = month_list
N_table['Day'] = day_list
N_table['Content'] = text_list 
N_table['Href_link'] = href_list 
N_table.to_csv('/Users/chenjessie/Desktop/News-crawler-master/BillAckman1.csv', index = False, header = True)
```
