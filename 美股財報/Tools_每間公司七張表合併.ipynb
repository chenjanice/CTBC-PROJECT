# 每間公司表合併（7張表）

```python
import glob
import pandas as pd
import csv
import os
```

```python
#建資料夾
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
createFolder('./Balance')
createFolder('./CashFlow')
createFolder('./Income')
```

```python
## 修改日期
def month_transform(month_string):
    q1 = ['Jan','Feb','Mar']
    q2 = ['Apr','May','Jun']
    q3 = ['Jul','Aug','Sep']
    q4 = ['Oct','Nov','Dec']
    
    if month_string in q1:
        return '01','01'
    elif month_string in q2:
        return '04','01'
    elif month_string in q3:
        return '07','01'
    elif month_string in q4:
        return '10','01'

#刪除空白列    
def deletespace(df):
    for i in range(df.shape[0]-1,-1,-1):
            if df.iloc[i][0] == '\xa0\xa0\xa0' :
                df = df.drop(index=i)
    return df
```

# Main

```python
file_type= ['*_Balance_Sheets_1.csv','*_Income_Statement_1.csv','*_CashFlow_Statements_1.csv']
for i in range(3):
    f_type = file_type[i].split('_')[1]
    for select in glob.glob('./CTBC/'+file_type[i]):#檔案路徑
        company_name = select.split('/')[-1].split('_')[0]
        df = pd.read_csv(select)
        print(select)
        df = deletespace(df)

        #合併檔案
        for i in range(2,8):
            try:
                path = select.split('1.csv')[0]+str(i)+'.csv'
                df1 = pd.read_csv(path)
                df1 = deletespace(df1)
                df = df.merge(df1,how='outer')
            except:
                continue

        #修改日期
        column_name = []
        for index in range(len(df.columns)):
            if index == 0:
                column_name.append(str(index))
            else:
                old_date = df.columns[index]
                month,day = month_transform(old_date.split('-')[0])
                year = old_date.split('-')[2]
                if '.' in year:
                    year = year.split('.')[0]
                new_date = '20'+year +'-'+ month +'-'+ day
                if '.' in new_date :
                    new_date = new_date.split('.')[0]
                column_name.append(new_date)
        df.columns = column_name
        #輸出成三個資料夾：f_type = Balance/CashFlow/Income(要先建好)
        df.to_csv('./'+ f_type +'/'+ company_name +'.csv', index=True)
```
