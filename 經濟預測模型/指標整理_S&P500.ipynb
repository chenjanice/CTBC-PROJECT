# 月報酬：[((Pt+1)-Pt)/Pt]**12
Pt+1 : 每個月一號的指數   

Pt：下個月一號的指數

```python
import datetime
x=(datetime.date.today())
for i in range(1,121):
    if x.month > 1 :
        x=x.replace(month=x.month-1,day=1)
    else:
        x=x.replace(year=x.year-1,month=12,day=1)
    print(x)
```

```python
import pandas as pd
df1= pd.read_csv('/Users/chenjessie/Desktop/經濟預測模型/s&p500.csv')
df1.head()
```

```python
df1['sp500'] = 0
df1['condition'] = 0
df1['predict'] = 0
for i in range(df1.shape[0]-1):
    rate = float(format(((df1.iloc[i+1]['Close'] - df1.iloc[i]['Close'])/ df1.iloc[i]['Close']),'.10f') )
    df1.loc[i,'sp500'] = rate
    
    if rate > 0.07:
        df1.loc[i,'condition'] = '3'
        if i-1>=0:
            df1.loc[i-1,'predict'] = '3'
    elif rate < (-0.07):
        df1.loc[i,'condition'] = '1'
        if i-1>=0:
            df1.loc[i-1,'predict'] = '1'
    else:
        df1.loc[i,'condition'] = '2'
        if i-1>=0:
            df1.loc[i-1,'predict'] = '2'
```

```python
df1=df1.drop([1109,1110])
df1
```

```python
df2 = pd.DataFrame(df1,columns=['Date','predict'])
df2
```

```python
df2.to_csv('example.csv', index=False)
```

# 處理日期問題

# [1] 日期換成  xxxx-xx-xx  格式

```python
df=pd.read_csv('/Users/chenjessie/Desktop/經濟預測模型/美国新屋销售总数年化.csv')
df=df.drop(columns=['前值(万户)'])
df.columns = ['日期','新屋銷售']
df
```

```python
for i in range(df.shape[0]):
    y = df['日期'].iloc[i].split('年')[0]
    m = df['日期'].iloc[i].split('年')[1].split('月')[0]
    date = y+'-'+m+'-01'
    print(date)
    df['日期'].iloc[i]=date
```

```python
df
```

# [2] 刪除重複日期，取第一筆

```python
df=df.drop_duplicates('日期','first')
df
```

# [3] 將日期設為index

```python
df['日期'] = pd.to_datetime(df['日期'])
df = df.set_index('日期') 
df
```

# [4] 創一個完整日期的pd

```python
import datetime
from dateutil.relativedelta import relativedelta
dic={}
dic['日期']=[]
in_date = '2020-04-01'
dt = datetime.datetime.strptime(in_date, "%Y-%m-%d")
for i in range(50*12+4):
    dt = datetime.datetime.strptime(in_date, "%Y-%m-%d")
    dt = (dt-relativedelta(months=i)).strftime("%Y-%m-%d")
    dic['日期'].append(dt)
    #print(dt)
    
#print(dic['Date'])
new = pd.DataFrame(dic)
new['日期'] = pd.to_datetime(new['日期'])
new = new.set_index('日期') 
new
```

# [5] Left Join

```python
res = pd.merge(new,df, on='日期',how='outer')
res
```

# [6] 用平均數填補空值

```python
print(res.isna().sum())
res.mean(column = 1) 
print(res.isna().sum())
```

# [7] 輸出

```python
res.to_csv('新屋銷售.csv', index=False)
```
